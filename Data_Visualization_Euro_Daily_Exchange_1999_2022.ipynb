```python
# Importing required packages
import pandas as pd
```

```python
#Reading data into DataFrame
df=pd.read_csv('./euro-daily-hist_1999_2022.csv')
```

```python
#Reading first five top rows
df.head(5)
```

```python
#Reading last five top rows
df.tail(5)
```

```python
#To know shape of the data (number of rows and columns)
print(f"No. of rows in train are {df.shape[0]} and No. of columns in train are {df.shape[1]}")
```

```python
# Lets see the dtypes and null value distributions of all columns
df.info()
```

```python
#To find null values in the dataset
df.isnull().sum()
```

```python
#Finding datatypes of each column
df.dtypes
```

```python
#Renaming columns for something easier to refer to
df.rename(columns={'[US dollar ]': 'US_dollar',
                    'Period\\Unit:': 'Time'},
                      inplace=True)
```

```python
#We change the Time column to a datetime data type
df.Time=pd.to_datetime(df.Time)
df.dtypes
```

```python
#Sort Values by Time 
df.sort_values('Time', inplace=True)
```

```python
#We reset the index (and drop the initial index)
df.reset_index(drop=True,inplace=True)
```

```python
#Creating New dataset with Time and US_Dollar
euro_to_dollar=df[['Time','US_dollar']].copy()
```

```python
euro_to_dollar.shape
```

```python
euro_to_dollar['US_dollar'].value_counts().nunique()
euro_to_dollar['US_dollar'].unique()
```

```python
#excluding all the rows where the '-' character appears in the US_dollar column
euro_to_dollar.drop(euro_to_dollar.index[euro_to_dollar['US_dollar'] == '-'], inplace=True)
# euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
```

```python
euro_to_dollar.shape
```

```python
#Convert the US_dollar column to a float data type
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype('float')
```

```python
euro_to_dollar.dtypes
```

```python
#30 days rolling mean
euro_to_dollar['rolling_mean']=euro_to_dollar['US_dollar'].rolling(30).mean()
```

```python
##Lets make plots 
import matplotlib.pyplot as plt
%matplotlib inline 

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

```python
plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original values', weight='bold')

for i, rolling_mean in zip([2, 3, 4, 5, 6],
                           [7, 30, 50, 100, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'],
             euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling Window:' + str(rolling_mean), weight='bold')
    
plt.tight_layout() # Auto-adjusts the padding between subplots
plt.show()
```

```python

#generating a line plot for Rolling_mean=30
fig,ax = plt.subplots(figsize=(10,5))
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['rolling_mean'])
plt.title('EURO - USD rate change between 1999 and 2021 with 30days Moving Average',size =16)
plt.xlabel('Year',size = 12)
plt.ylabel('Rate',size = 12)
plt.show()
```

 The graph above represents of 30 days moving average of daily exchange rate of Euro-USD.

# Euro-dollar rate change during Coronavirus PandemicÂ¶

Let's have closer look to the more specific period of time and explore how the euro-dollar rate has changed during the Coronavirus Pandemic. The 2020 data will be shown and the 2016-2019 data will be used as a baseline.

```python
#preparing the required time frame
pandemic_euro_baseline=euro_to_dollar.copy()[
                     (euro_to_dollar['Time'].dt.year >= 2016) &
                     (euro_to_dollar['Time'].dt.year <= 2021)  ]
pandemic_euro_2020=euro_to_dollar.copy()[
                     (euro_to_dollar['Time'].dt.year >= 2020) &
                     (euro_to_dollar['Time'].dt.year <= 2021) ]
                      
```

```python
# adding the style
import matplotlib.style as style
style.use('fivethirtyeight')

#generating a line plot
fig,ax = plt.subplots(figsize=(10,5))

ax.plot(pandemic_euro_baseline['Time'],
        pandemic_euro_baseline['rolling_mean'],
        linewidth=1,color='blue')

# adding the highlight part for pandemic period
ax.plot(pandemic_euro_2020['Time'],
        pandemic_euro_2020['rolling_mean'],
        linewidth=3,color='red')

#changing the color of tick labels
ax.tick_params(axis ='x', colors = 'grey')
ax.tick_params(axis ='y', colors = 'grey')

ax.set_title('EURO - USD rate change during 2016-2021 period of time with Main focus on Pandemic period')

```

# Euro-Dollar rate fluctuations during "The Three US Presidencies ""

```python
bush_obama_trump = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2001) & (euro_to_dollar['Time'].dt.year < 2021)]
bush = bush_obama_trump.copy(
       )[bush_obama_trump['Time'].dt.year < 2009]
obama = bush_obama_trump.copy(
       )[(bush_obama_trump['Time'].dt.year >= 2009) & (bush_obama_trump['Time'].dt.year < 2017)]
trump = bush_obama_trump.copy(
       )[(bush_obama_trump['Time'].dt.year >= 2017) & (bush_obama_trump['Time'].dt.year < 2021)]
```

```python
### Adding the FiveThirtyEight style
style.use('fivethirtyeight')

### Adding the subplots
plt.figure(figsize=(12, 6))
ax1 = plt.subplot(2,3,1)
ax2 = plt.subplot(2,3,2)
ax3 = plt.subplot(2,3,3)
ax4 = plt.subplot(2,1,2)
axes = [ax1, ax2, ax3, ax4]

for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2','1.4', '1.6'],
                   alpha=0.3)
    ax.grid(alpha=0.5)    
    

### Ax1: Bush
ax1.plot(bush['Time'], bush['rolling_mean'],color='purple')
ax1.set_xticklabels(['', '2001', '', '2003', '', '2005', '',
                     '2007', '', '2009'],
                   alpha=0.3)
ax1.set_title("BUSH 2001-2009",fontsize=18, weight='bold',color='purple')


### Ax2: Obama
ax2.plot(obama['Time'], obama['rolling_mean'],color='orange')
ax2.set_xticklabels(['', '2009', '', '2011', '', '2013', '',
                     '2015', '', '2017'], alpha=0.3)
ax2.set_title("OBAMA 2009-2017",fontsize=18, weight='bold',color='orange')


### Ax3: Trump
ax3.plot(trump['Time'], trump['rolling_mean'],color='green')
ax3.set_xticklabels(['2017', '', '2018', '', '2019', '',
                     '2020', '', '2021'],
                   alpha=0.3)
ax3.set_title("TRUMP 2017-2021",fontsize=18, weight='bold',color='green')


### Ax4: Bush-Obama-Trump
ax4.plot(bush['Time'], bush['rolling_mean'],
        color='purple')
ax4.plot(obama['Time'], obama['rolling_mean'],
        color='orange')
ax4.plot(trump['Time'], trump['rolling_mean'],
        color='green')
ax4.grid(alpha=0.5)
ax4.set_xticks([])


#ax4.set_title("EURO-USD Exchange rates during last three US presidencies",fontsize=24, weight='bold',color='black')
```
